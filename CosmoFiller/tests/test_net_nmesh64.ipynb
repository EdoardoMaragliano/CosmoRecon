```python
field_size = 64
mask_channels = 1
```

```python
base_model_obj = InpaintingModel(input_field='rho', norm_val=20.0)
base_model_obj.set_logger(logging)

base_model = base_model_obj.prepare_model(
    input_size=(field_size, field_size, field_size, mask_channels)
    )

# masked_model = MaskedInpaintingModel(inputs=base_model.input, outputs=base_model.output)
base_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4, amsgrad=True, global_clipnorm=1e-2),
    loss=["mean_squared_error"],
    metrics=["mean_squared_error"],
    )

logging.info("Model summary:\n%s", str(base_model.summary()))
```
