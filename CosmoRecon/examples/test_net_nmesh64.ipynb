# Model Build Test (64³ Grid)

Builds and compiles the 3D U-Net at mesh size 64 and prints the model summary
to inspect the architecture and parameter count.

```python
import logging
import sys

from tensorflow import keras
from CosmoRecon.models import UNet
```

```python
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[logging.StreamHandler(sys.stdout)],
    force=True,
)
```

```python
field_size = 64
mask_channels = 1

builder = UNet(input_field="rho", norm_val=20.0)
builder.set_logger(logging)

model = builder.prepare_model(
    input_size=(field_size, field_size, field_size, mask_channels)
)

model.compile(
    optimizer=keras.optimizers.Adam(
        learning_rate=1e-4, amsgrad=True, global_clipnorm=1e-2
    ),
    loss="mean_squared_error",
    metrics=["mean_squared_error"],
)

model.summary()
```
